```python
import pandas as pd

data={
    'Region':['North','South','East','West','North','South','East','West'],
    'Product':['A','B','A','B','C','C','B','A'],
    'Sales':[150,200,300,400,250,180,220,310],
    'Quantity':[10,15,20,25,12,14,16,18],
}
df=pd.DataFrame(data)
print("Sample Dataset:\n",df)
```

```python
sales_by_region_sonal=df.groupby('Region')['Sales'].sum()
print("\nTotal Sales by Region:\n",sales_by_region_sonal)
```

```python
mean_by_product_sonal=df.groupby('Product')[['Sales','Quantity']].mean()
print("\nMean Sales and Quantity by Product:\n",mean_by_product_sonal)
```

```python
count_by_region_sonal=df.groupby('Region')['Sales'].count()
print("\nCount of Sales Records by Region:\n",count_by_region_sonal)
```

```python
custom_aggregation_sonal=df.groupby('Region')['Sales'].agg(['min','max'])
print("\nCustom Aggregation:\n",custom_aggregation_sonal)
```

```python
multi_level_agg_sonal=df.groupby(['Region','Product'])['Sales'].sum()
print("\nSales by Region and Product:\n",multi_level_agg_sonal)
```

```python
multi_level_agg_reset_sonal=multi_level_agg_sonal.reset_index()
print("\nSales by Region and Product:\n", multi_level_agg_reset_sonal)
```

```python
import pandas as pd

data_sonal={
    'Region':['North','South','East','West','North','South','East','West'],
    'City': ['City1', 'City2', 'City3', 'City4', 'City1', 'City2', 'City3', 'City4'],
    'Product':['A','B','A','B','C','C','B','A'],
    'Sales':[150,200,300,400,250,180,220,310],
    'Quantity':[10,15,20,25,12,14,16,18],
    'Date':pd.to_datetime(['2024-01-01', '2024-01-02', '2024-02-01', '2024-02-03',
                            '2024-03-01', '2024-03-02', '2024-04-01', '2024-04-03'])
}
df_sonal=pd.DataFrame(data_sonal)
print("Extended Dataset:\n",df_sonal)
```

```python
df_sonal.set_index('Date', inplace=True)
```

```python
monthly_sales_sonal=df_sonal.resample('M')['Sales'].sum()
print("\nTotal Sales by Month:\n",monthly_sales_sonal)
```

```python
quarterly_sales_sonal = df_sonal.resample('Q')['Sales'].sum()
print("\nTotal Sales by Quarter:\n", quarterly_sales_sonal)
```

```python
yearly_sales_sonal=df_sonal.resample('Y')['Sales'].sum()
print("\nTotal Sales by Year:\n",yearly_sales_sonal)
```

```python
df_sonal.reset_index(inplace=True)
```

```python
sales_by_region_sonal = df_sonal.groupby('Region')['Sales'].sum()
print("\nTotal Sales by Region:\n", sales_by_region_sonal)
```

```python
sales_by_city_sonal= df_sonal.groupby('City')['Sales'].sum()
print("\nTotal Sales by City:\n", sales_by_city_sonal)
```

```python
sales_by_region_city_sonal=df_sonal.groupby(['Region','City'])['Sales'].sum()
print("\nTotal Sales by Region and City:\n",sales_by_region_city_sonal)
```

```python
sales_by_region_city_reset_sonal=sales_by_region_city_sonal.reset_index()
sales_by_region_city_reset_sonal.to_csv("spatial_aggregation.csv", index=False)
print("\nSpatial aggregation data saved to 'spatial_aggregation.csv'")
  
```
